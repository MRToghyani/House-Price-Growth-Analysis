```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2023 = pd.read_csv("2023_v4_Area.csv")
df2021 = pd.read_csv("2021_v4_Area.csv")
s = pd.read_csv("supplementary_table")
```

```python
df2021
```

```python
df2021.describe()
```

```python
df20211 = df2021.copy()
```

```python
df20211["Area"] = pd.cut(df20211["Area"] , bins = [0,60,120,200] ) #, labels= ["small" , "med" , "big"]
df20211
```

```python
df20211["Construction"] = pd.cut(df20211["Construction"] , bins = [1370,1380,1390,1400] , include_lowest=True ) #, labels= ["small" , "med" , "big"]
df20211
```

```python
def rep(x):
    if x in s.District.values:
        return s[s.District == x]["Profit"].values[0]
    else:
        return np.nan
```

```python
df20211["District"] = df20211.District.apply(lambda x : rep(x))
```

```python
df20212 = df20211.dropna().reset_index(drop = True)
df20212
```

```python
g = df20211.groupby(["Area","Construction","District"]).agg( Price_mean = pd.NamedAgg(column= "Price" , aggfunc="mean"))
f2021 = g.reset_index(drop = False)
f2021
```

```python
df20231 = df2023.copy()
```

```python
df20231["Area"] = pd.cut(df20231["Area"] , bins = [0,60,120,200] ) #, labels= ["small" , "med" , "big"]
df20231
```

```python
df20231["Construction"] = pd.cut(df20231["Construction"] , bins = [1370,1380,1390,1400] , include_lowest=True ) #, labels= ["small" , "med" , "big"]
df20231
```

```python
df20231["District"] = df20231.District.apply(lambda x : rep(x))
```

```python
df20232 = df20211.dropna().reset_index(drop = True)
df20232
```

```python
g = df20231.groupby(["Area","Construction","District"]).agg( Price_mean = pd.NamedAgg(column= "Price" , aggfunc="mean"))
f2023 = g.reset_index(drop = False)
f2023
```

```python
m20212023 = pd.merge(f2021,f2023 , on =["Area","Construction","District"] , how = "inner", suffixes=["_2021","_2023"])
m20212023
```

```python
m20212023["Profit"] = (m20212023.Price_mean_2023 - m20212023.Price_mean_2021) / m20212023.Price_mean_2021
m20212023
```

```python
m20212023.sort_values(by = "Profit", ascending=False,inplace= True)
m20212023
```

```python
f = m20212023.dropna().reset_index(drop = True)
f
```

```python
f.info()
```

```python
f.Area=f.Area.astype(str)
f.Construction=f.Construction.astype(str)
f.District=f.District.astype(str)
```

```python
f["description"] ="Area : "+ f.Area + " ,Construction : " + f.Construction + " ,District : " + f.District
```

```python
f
```

```python
viz = f.copy()
viz["Year_1"] = "2021"
viz["Year_2"] = "2023"
viz_f = pd.DataFrame({"Year" : pd.concat([viz["Year_1"],viz["Year_2"]]) , 
                                 "Price_mean" : pd.concat([viz["Price_mean_2021"],viz["Price_mean_2023"]]),
                                 "description" : pd.concat([viz["description"],viz["description"]])  })
```

```python
display(m20212023.reset_index(drop =True))
sns.pointplot(viz_f, x="Year", y="Price_mean")
plt.show()
ax = sns.pointplot(viz_f, x="Year", y="Price_mean" , hue= "description",log_scale =True ,dodge= 0.01,legend="full")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```
