```python
df20211 = df2021.copy()
```

```python
df20211["Area"] = pd.cut(df20211['Area'] , bins= [0,20,40,60,80,100,120,140,160,180,200])
```

```python
df20211
```

```python
g = df20211.groupby("Area")["Price"].mean()
f2021 = pd.DataFrame(g)
f2021 = f2021.reset_index(drop = False)
f2021
```

```python
df20231 = df2023.copy()
```

```python
df20231["Area"] = pd.cut(df20231['Area'] , bins= [0,20,40,60,80,100,120,140,160,180,200] , include_lowest=False)
```

```python
df20232 = df20231.dropna(subset = "Area").reset_index(drop =True)
df20232
```

```python
g2023 = df20232.groupby("Area")["Price"].mean()
f2023 = pd.DataFrame(g2023)
f2023 = f2023.reset_index()
f2023
```

```python
m20212023 = pd.merge(f2021,f2023,how= "inner", on= "Area",suffixes=["_2021","_2023"])
m20212023
```

```python
m20212023["Profit"] = (m20212023.Price_2023 - m20212023.Price_2021) / m20212023.Price_2021
m20212023
```

```python
m20212023.sort_values(by = "Profit", ascending=False,inplace= True)
m20212023
```

```python
viz = m20212023.copy()
viz["Year_1"] = "2021"
viz["Year_2"] = "2023"
viz_f = pd.DataFrame({"Year" : pd.concat([viz["Year_1"],viz["Year_2"]]) , 
                                 "Price_mean" : pd.concat([viz["Price_2021"],viz["Price_2023"]]),
                                 "Area_range" : pd.concat([viz["Area"],viz["Area"]])  })
```

```python
display(m20212023)
sns.pointplot(viz_f, x="Year", y="Price_mean")
plt.show()
ax = sns.pointplot(viz_f, x="Year", y="Price_mean" , hue= "Area_range",log_scale =True ,dodge= 0.01,legend="full")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
ax = sns.catplot(data =m20212023, x= "Area" , y="Profit", native_scale = True ,kind="point" )
ax.set_xticklabels( rotation=40, ha="right")
#plt.tight_layout()
plt.show()
```

```python
df2023 = pd.read_csv("2023_v4.csv")
df2021 = pd.read_csv("2021_v4.csv")
```

```python
df2021
```

```python
df20211 = df2021.replace(True ,1)
df20211 = df20211.replace(False ,0)
```

```python
df20211
```

```python
df20211["Features"] = df20211.Warehouse + df20211.Parking + df20211.Elevator
df20211
```

```python
g = df20211.groupby("Features")["Price_adj"].mean()
f2021 = pd.DataFrame(g).reset_index()
f2021
```

```python
df2023
```

```python
df20231 = df2023.replace(True ,1)
df20231 = df20231.replace(False ,0)
df20231
```

```python
df20231["Features"] = df20231.Warehouse + df20231.Parking + df20231.Elevator
df20231
```

```python
g = df20231.groupby("Features")["Price_adj"].mean()
f2023 = pd.DataFrame(g).reset_index()
f2023
```

```python
f2021
```

```python
m20212023 = pd.merge(f2021,f2023,how= "inner", on= "Features",suffixes=["_2021","_2023"])
m20212023
```

```python
m20212023["Profit"] = (m20212023.Price_adj_2023 - m20212023.Price_adj_2021) / m20212023.Price_adj_2021
m20212023
```

```python
m20212023.sort_values(by = "Profit",ascending=False)
```

```python
viz = m20212023.copy()
viz["Year_1"] = "2021"
viz["Year_2"] = "2023"
viz_f = pd.DataFrame({"Year" : pd.concat([viz["Year_1"],viz["Year_2"]]) , 
                                 "Price_mean" : pd.concat([viz["Price_adj_2021"],viz["Price_adj_2023"]]),
                                 "Features" : pd.concat([viz["Features"],viz["Features"]])  })
```

```python
display(m20212023)
sns.pointplot(viz_f, x="Year", y="Price_mean")
plt.show()
ax = sns.pointplot(viz_f, x="Year", y="Price_mean" , hue= "Features",log_scale =True ,dodge= 0.01,legend="full")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```
