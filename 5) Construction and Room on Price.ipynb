```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2023 = pd.read_csv("2023_v4.csv")
df2021 = pd.read_csv("2021_v4.csv")
```

```python
df20211 = df2021.copy()
```

```python
np.linspace(1370,1400,num =7)
```

```python
df20211["Construction"] = pd.cut(df20211['Construction'] , bins= [1370,1375,1380,1385,1390,1395,1400],include_lowest=True)
df20211
```

```python
g = df20211.groupby("Construction")["Price_adj"].mean()
f2021 = pd.DataFrame(g)
f2021 = f2021.reset_index(drop = False)
f2021
```

```python
df20231 = df2023.copy()
```

```python
df20231["Construction"] = pd.cut(df20231['Construction'] , bins= [1370,1375,1380,1385,1390,1395,1400],include_lowest=True)
df20231
```

```python
g2023 = df20231.groupby("Construction")["Price_adj"].mean()
f2023 = pd.DataFrame(g2023)
f2023 = f2023.reset_index()
f2023
```

```python
m20212023 = pd.merge(f2021,f2023,how= "inner", on= "Construction",suffixes=["_2021","_2023"])
m20212023
```

```python
m20212023["Profit"] = (m20212023.Price_adj_2023 - m20212023.Price_adj_2021) / m20212023.Price_adj_2021
m20212023
```

```python
m20212023_1 = m20212023.sort_values(by = "Profit", ascending=False,inplace= False)
m20212023_1
```

```python
viz = m20212023_1.copy()
viz["Year_1"] = "2021"
viz["Year_2"] = "2023"
viz_f = pd.DataFrame({"Year" : pd.concat([viz["Year_1"],viz["Year_2"]]) , 
                                 "Price_mean" : pd.concat([viz["Price_adj_2021"],viz["Price_adj_2023"]]),
                                 "Construction_range" : pd.concat([viz["Construction"],viz["Construction"]])  })
```

```python
display(m20212023_1)
sns.pointplot(viz_f, x="Year", y="Price_mean",log_scale =True)
plt.show()
ax = sns.pointplot(viz_f, x="Year", y="Price_mean" , hue= "Construction_range",log_scale =True ,dodge= 0.1,legend="full")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
ax = sns.catplot(data =m20212023, x= "Construction" , y="Profit", native_scale = True ,kind="point" )
ax.set_xticklabels( rotation=40, ha="right")
#plt.tight_layout()
plt.show()
```

```python
m20212023["Profit"] = (m20212023.Price_adj_2023 - m20212023.Price_adj_2021) / m20212023.Price_adj_2021
m20212023
```

```python
viz = m20212023.copy()
viz["Year_1"] = "2021"
viz["Year_2"] = "2023"
viz_f = pd.DataFrame({"Year" : pd.concat([viz["Year_1"],viz["Year_2"]]) , 
                        "Price_mean" : pd.concat([viz["Price_adj_2021"],viz["Price_adj_2023"]]),
                         "Age_range" : pd.concat([viz["Age"],viz["Age"]])  })
```
