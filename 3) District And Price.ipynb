```python
g2021 = df2021.groupby("District").agg(
    Price_adj = pd.NamedAgg(column="Price_adj" , aggfunc="mean"),
    count = pd.NamedAgg(column="Price_adj" , aggfunc="count"))
gdf2021 = g2021.reset_index(drop=False)
gdf2021
```

```python
# g2023 = df2023.groupby("District")["Price_adj"].mean()
# gdf2023 = pd.DataFrame(g2023)
# gdf2023 = gdf2023.reset_index(drop=False)
# gdf2023
```

```python
g2023 = df2023.groupby("District").agg(
    Price_adj = pd.NamedAgg(column="Price_adj" , aggfunc="mean"),
    count = pd.NamedAgg(column="Price_adj" , aggfunc="count"))
gdf2023 = g2023.reset_index(drop=False)
gdf2023
```

```python
mdf = pd.merge(gdf2021,gdf2023, on=["District"], how='inner',suffixes=["_2021","_2023"])
mdf
```

```python
mdf["Profit"] = (mdf.Price_adj_2023 - mdf.Price_adj_2021) / mdf.Price_adj_2021
```

```python
mdf.sort_values(by = "Profit", inplace=True)
mdf = mdf.reset_index(drop = True)
mdf
```

```python
mean_2021 = mdf.Price_adj_2021.mean()
mean_2023 = mdf.Price_adj_2023.mean()
Overal_profit = (mean_2023 - mean_2021) / mean_2021
print("mean 2021 price : ",mean_2021)
print("mean 2021 price : ",mean_2023)
print("Overal profit : ",Overal_profit)
```

```python
sns.displot(mdf, x="Profit")
```

```python
profit = np.asanyarray(mdf.Profit)
profit
```

```python
profit = profit.reshape(-1, 1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

data_transformed = MinMaxScaler().fit_transform(profit)
data_transformed
```

```python
Sum_of_squared_distances = []
K = range(1,30)
for k in K:
    km = KMeans(init = "k-means++", n_clusters = k, n_init = 12)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# optimum = 12
km = KMeans(init = "k-means++", n_clusters = 12, n_init = 12)
km = km.fit(data_transformed)
y = km.predict(data_transformed)
y
```

```python
mdf["Group"] = y
```

```python
mdf2 =mdf .copy()
```

```python
mdf2["Sum_2021"] = mdf2.Price_adj_2021 * mdf2.count_2021
mdf2["Sum_2023"] = mdf2.Price_adj_2023 * mdf2.count_2023
```

```python
g = mdf2.groupby("Group")
groups_df = pd.DataFrame({"Price_mean_2021" :g.Sum_2021.sum() / g.count_2021.sum(),
                 "Price_mean_2023" :g.Sum_2023.sum() / g.count_2023.sum()})
groups_df
```

```python
groups_df["Group_profit"] = (groups_df.Price_mean_2023 - groups_df.Price_mean_2021) / groups_df.Price_mean_2021
groups_df
```

```python
groups_df.sort_values(by = "Group_profit" , ascending=False, inplace=True)
groups_df
```
